```python
#Task 5 – Votes Analysis (Zomato Dataset)

#This task generally includes:

#Find the restaurant with the highest votes.
#Find the restaurant with the lowest votes.
#Analyze the relationship between votes and ratings.
#Visualize the relationship using a scatter plot.
```

```python
import pandas as pd

df = pd.read_csv("Zomato-data-.csv")

df.head()
```

```python
#Step 2: Restaurant with Highest Votes
```

```python
highest_votes = df.loc[df['votes'].idxmax()]

print("Restaurant with Highest Votes:")
print(highest_votes)
```

```python
#Step 3: Restaurant with Lowest Votes
```

```python
lowest_votes = df.loc[df['votes'].idxmin()]

print("Restaurant with Lowest Votes:")
print(lowest_votes)
```

```python
#Step 4: Correlation Between Votes and Ratings

#If your rate column contains /5, first clean it:
```

```python
df['rate'] = df['rate'].astype(str).str.replace('/5', '', regex=False)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
correlation = df[['votes', 'rate']].corr()

print(correlation)
```

```python
#Step 5: Scatter Plot
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(df['votes'], df['rate'])

plt.xlabel("Votes")
plt.ylabel("Ratings")
plt.title("Relationship Between Votes and Ratings")

plt.show()
```

```python
#Report Explanation

#Objective:
#To analyze restaurant popularity using the number of votes and examine its relationship with ratings.

#Analysis Performed:

#Identified restaurants with the maximum and minimum votes.
#Calculated correlation between votes and ratings.
#Created a scatter plot to visualize the relationship.

#Conclusion:
#Restaurants with higher ratings generally tend to receive more votes, indicating a positive relationship between customer ratings and popularity.
```
